* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ – 3.

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
import time

def prod_non_zero_diag(x):
    """Compute product of nonzero elements from matrix diagonal.

    input:
    x -- 2-d numpy array
    output:
    product -- integer number


    Not vectorized implementation.
    """
    ans = 1
    n = len(x)
    m = len(x[0])
    
    i = 0
    j = 0
    while i < n and j < m:
        if x[i][j] != 0:
            ans *= x[i][j]
        i += 1
        j += 1

    return ans

    pass

def prod_non_zero_diag_np(x):
    """Compute product of nonzero elements from matrix diagonal.

    input:
    x -- 2-d numpy array
    output:
    product -- integer number


    Vectorized implementation.
    """
    bro = np.diagonal(x)
    zxc = bro[bro != 0]
    return np.prod(zxc)
    
    pass

def draw(data_size, time_vectorized, time_non_vectorized):
    f, ax1 = plt.subplots()
    ax1.plot(data_size, time_vectorized, label=u"Векторизованная")
    ax1.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
    ax1.set_xlabel(u"Длина массива")
    ax1.set_ylabel(u"Время работы(с)")
    ax1.grid()
    ax1.legend()
    plt.show()


data_size = [0] * 400
time_non_vectorized = [0] * 400
time_vectorized = [0] * 400

for i in range(400):
    data_size[i] = rnd.randint(100, 1000)
data_size.sort()

for i in range(400):
    n = int(data_size[i])
    m = n + 1
    
    a = [[rnd.randint(-50, 50) for _ in range(m)] for _ in range(n)]
    
    st_time = time.time()
    res = prod_non_zero_diag(a)
    end_time = time.time()
    time_non_vectorized[i] = end_time - st_time
    
    a_np = np.array(a)
    
    st_time = time.time()
    res_np = prod_non_zero_diag_np(a_np)
    end_time = time.time()
    time_vectorized[i] = end_time - st_time
draw(data_size, time_vectorized, time_non_vectorized)
    
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
import time

def draw(data_size, time_vectorized, time_non_vectorized):
    f, ax1 = plt.subplots()
    ax1.plot(data_size, time_vectorized, label=u"Векторизованная")
    ax1.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
    ax1.set_xlabel(u"Длина массива")
    ax1.set_ylabel(u"Время работы(с)")
    ax1.grid()
    ax1.legend()
    plt.show()

def are_multisets_equal_np(x, y):
    """Return True if both vectors create equal multisets.

    input:
    x, y -- 1-d numpy arrays
    output:
    True if multisets are equal, False otherwise -- boolean

    Not vectorized implementation.
    """
    x = np.sort(x)
    y = np.sort(y)
    
    return np.array_equal(x, y)
    
    pass

def are_multisets_equal(x, y):
    """Return True if both vectors create equal multisets.

    input:
    x, y -- 1-d numpy arrays
    output:
    True if multisets are equal, False otherwise -- boolean

    Not vectorized implementation.
    """
    x.sort()
    y.sort()
    
    n = len(x)
    m = len(y)
    
    if n != m:
        return False
    for i in range(n):
        if x[i] != y[i]:
            return False
    return True

    pass

data_size = [0] * 600
time_non_vectorized = [0] * 600
time_vectorized = [0] * 600
for i in range(600):
    data_size[i] = rnd.randint(20, 100000)
data_size.sort()
for i in range(600):
    n = data_size[i]
    
    a = [rnd.randint(-40, 40) for _ in range(n)]
    b = [rnd.randint(-40, 40) for _ in range(n)]
    st_time = time.time()
    res = are_multisets_equal(a, b)
    time_non_vectorized[i] = time.time() - st_time
    
    a = np.array(a)
    b = np.array(b)
    st_time = time.time()
    res_np = are_multisets_equal_np(a, b)
    time_vectorized[i] = time.time() - st_time
    
draw(data_size, time_vectorized, time_non_vectorized)
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит нулевой.  
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ – 5.

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
import time

def draw(data_size, time_vectorized, time_non_vectorized):
    f, ax1 = plt.subplots()
    ax1.plot(data_size, time_vectorized, label=u"Векторизованная")
    ax1.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
    ax1.set_xlabel(u"Длина массива")
    ax1.set_ylabel(u"Время работы(с)")
    ax1.grid()
    ax1.legend()
    plt.show()

def max_after_zero_np(x):
    """Find max element after zero in array.

    input:
    x -- 1-d numpy array
    output:
    maximum element after zero -- integer number

    Not vectorized implementation.
    """
    bro = x == 0

    bro = np.roll(bro, 1)
    bro[0] = False
    ans = np.max(x, where=bro, initial=0)

    return ans

    pass
    
def max_after_zero(x):
    """Find max element after zero in array.

    input:
    x -- 1-d numpy array
    output:
    maximum element after zero -- integer number

    Not vectorized implementation.
    """
    
    ans = 0
    
    for i in range(1, len(x)):
        if x[i - 1] == 0:
            ans = max(ans, x[i])
            
    return ans

    pass
    
data_size = [0] * 600
time_non_vectorized = [0] * 600
time_vectorized = [0] * 600
for i in range(600):
    data_size[i] = rnd.randint(100, 100000)
data_size.sort()
for i in range(600):
    n = data_size[i]
    
    a = [rnd.randint(-100, 100) for _ in range(n)]
    st_time = time.time()
    res = max_after_zero(a)
    time_non_vectorized[i] = time.time() - st_time
    
    a = np.array(a)
    st_time = time.time()
    res = max_after_zero_np(a)
    time_vectorized[i] = time.time() - st_time
    
draw(data_size, time_vectorized, time_non_vectorized)
```

* __Задача 5__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ – (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
import time

def draw(data_size, time_vectorized, time_non_vectorized):
    f, ax1 = plt.subplots()
    ax1.plot(data_size, time_vectorized, label=u"Векторизованная")
    ax1.plot(data_size, time_non_vectorized, label=u"Невекторизованная")
    ax1.set_xlabel(u"Длина массива")
    ax1.set_ylabel(u"Время работы(с)")
    ax1.grid()
    ax1.legend()
    plt.show()

def run_length_encoding_np(x):
    """Make run-length encoding.

    input:
    x -- 1-d numpy array
    output:
    elements, counters -- integer iterables

    Vectorized implementation.
    """
    x = np.append(x, x[-1] + 1) #добавляем фиктивный чтобы не пропустить последний элемент
    d = np.diff(x)
    tmp = [-1] #добавляем фиктивный чтобы не пропустить первый элемент
    zxc = np.nonzero(d) #первые вхождения всех уникальных
    tmp = np.append(tmp, zxc)
    a = x[zxc] #все уникальные
    b = np.diff(tmp) #считаем кол-во вхождений
    return (a, b)

    pass


def run_length_encoding(x):
    """Make run-length encoding.

    input:
    x -- 1-d numpy array
    output:
    elements, counters -- integer iterables

    Not vectorized implementation.
    """
    a = []
    b = []
    c = 0

    prev = x[0]
    for el in x:
        if el == prev:
            c += 1
        else:
            a.append(prev)
            b.append(c)
            c = 1
        prev = el
    a.append(prev)
    b.append(c)

    return (a, b)
    pass

data_size = [0] * 600
time_non_vectorized = [0] * 600
time_vectorized = [0] * 600
for i in range(600):
    data_size[i] = rnd.randint(222, 100000)
data_size.sort()

for i in range(600):
    n = data_size[i]
    a = [rnd.randint(-30, 30) for _ in range(n)]
    
    st_time = time.time()
    (res1, res2) = run_length_encoding(a)
    time_non_vectorized[i] = time.time() - st_time
    
    a = np.array(a)
    st_time = time.time()
    (res1_np, res2_np) = run_length_encoding_np(a)
    time_vectorized[i] = time.time() - st_time
    
draw(data_size, time_vectorized, time_non_vectorized)
    
```
